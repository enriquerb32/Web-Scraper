```python
!pip install selenium
!apt-get update 
!apt install chromium-chromedriver

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait


chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver =webdriver.Chrome('chromedriver',chrome_options=chrome_options)
wait = WebDriverWait(driver, 10)
```

```python
keyword="iphone"
```

```python
# We go to the desired website and add the keyword which corresponds to our query
driver.get("https://www.amazon.es/s?k="+keyword)
```

```python
# We obtain the information thanks to Selenium fetching the source code
content = driver.page_source
```

```python
# Now we use BeautifulSoup to exploit the information
soup = BeautifulSoup(content)
```

```python
soup
```

```python
# We want to obtain the title, price and image of certain Amazon products
# Firstly, we inspect the source code of some products link, and retrieve the class besides the products ID
result = soup.findAll(class_ = "a-section a-spacing-base")
```

```python
result
```

```python
# It may happen that depending on the keyword, the webpage could change the class acompanying the product name
# We can try different classes that iterate over the source code assuming the condition that the previous statements did not fetch any product

if len(result) ==0:
    result = soup.findAll(class_ ="a-section a-spacing-medium a-text-center")
```

```python
df_list_products = pd.DataFrame(columns=["title","price","image"])
```

```python
# We create new directories on Drive with the name of that we are looking for
# We create new folders ensuring that the folders did not already exist

if not os.path.exists("/content/drive/MyDrive/Colab Notebooks/WS_"+keyword):
    os.makedirs("/content/drive/MyDrive/Colab Notebooks/WS_" +keyword)
```

```python
folder = "/content/drive/MyDrive/Colab Notebooks/WS_"+keyword+"/"
```

```python
# We iterate over our dataset and fetch the titles, prices and images

for element in result:
    title_product = element.find(class_= "a-size-base-plus a-color-base a-text-normal").get_text()
    price_product= element.find(class_= "a-price-whole")
    image_product = element.find("img")
    if price_product != None:   # If any product has no price, we consider it as a null value
        price_product= price_product.get_text()
    else:
        price_product= "Not existing price"

    # 'basename' reservates the name of a product (from source 'src') to write its image therein (with writing binary mode 'wb')
    fichero = open(folder+basename(image_product.attrs["src"]),"wb")   
    fichero.write(requests.get(image_product.attrs["src"]).content)

    df_list_products = df_list_products.append({"title":title_product, "price":price_product, "image":"<img src='"+image_product.attrs["src"]+"'>"},
    ignore_index=True)
```

```python
df_list_products.to_csv(folder+keyword+".csv")
```

```python
df_list_products.to_html(folder+keyword+".html", escape=False)
```
